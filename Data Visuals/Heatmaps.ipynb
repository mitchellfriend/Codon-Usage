```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
codon_usage = pd.read_csv('../EDA/Cleaned Codon Usage Data Set.csv')
```

```python
codon_usage.head()
```

```python
kingdom_list = []
for k,v in dict(codon_usage['Kingdom'].value_counts()).items():
    kingdom_list.append(k)
kingdom_list
```

```python
Codon_column_list = ['UUU','UUC', 'UUA', 'UUG', 'CUU', 'CUC', 'CUA', 'CUG', 'AUU', 'AUC', 'AUA',
                       'AUG', 'GUU', 'GUC', 'GUA', 'GUG', 'GCU', 'GCC', 'GCA', 'GCG', 'CCU',
                       'CCC', 'CCA', 'CCG', 'UGG', 'GGU', 'GGC', 'GGA', 'GGG', 'UCU', 'UCC',
                       'UCA', 'UCG', 'AGU', 'AGC', 'ACU', 'ACC', 'ACA', 'ACG', 'UAU', 'UAC',
                       'CAA', 'CAG', 'AAU', 'AAC', 'UGU', 'UGC', 'CAU', 'CAC', 'AAA', 'AAG',
                       'CGU', 'CGC', 'CGA', 'CGG', 'AGA', 'AGG', 'GAU', 'GAC', 'GAA', 'GAG',
                       'UAA', 'UAG', 'UGA']
```

```python
bacteria_df = codon_usage[codon_usage['Kingdom']=='bacteria'][Codon_column_list]
mitochondria_df = codon_usage[codon_usage['Kingdom']=='mitochondria'][Codon_column_list]
virus_df = codon_usage[codon_usage['Kingdom']=='virus'][Codon_column_list]
plant_df = codon_usage[codon_usage['Kingdom']=='plant'][Codon_column_list]
invertebrate_df = codon_usage[codon_usage['Kingdom']=='invertebrate'][Codon_column_list]
chloroplast_df = codon_usage[codon_usage['Kingdom']=='chloroplast'][Codon_column_list]
vertebrate_df = codon_usage[codon_usage['Kingdom']=='vertebrate'][Codon_column_list]
bacteriophage_df = codon_usage[codon_usage['Kingdom']=='bacteriophage'][Codon_column_list]
archaea_df = codon_usage[codon_usage['Kingdom']=='archaea'][Codon_column_list]
mammal_df = codon_usage[codon_usage['Kingdom']=='mammal'][Codon_column_list]
primate_df = codon_usage[codon_usage['Kingdom']=='primate'][Codon_column_list]
rodent_df = codon_usage[codon_usage['Kingdom']=='rodent'][Codon_column_list]
plasmid_df = codon_usage[codon_usage['Kingdom']=='plasmid'][Codon_column_list]
```

```python
fig,(ax1,ax2, ax3, ax4, ax5) = plt.subplots(5, 1, figsize=(10,10))

g1 = sns.heatmap(mitochondria_df, cbar=False, ax=ax1)
ax1.title.set_text('mitochondria')
g2 = sns.heatmap(mammal_df, cbar=False, ax=ax2)
ax2.title.set_text('mammal')
g3 = sns.heatmap(primate_df, cbar=False, ax=ax3)
ax3.title.set_text('primate')
g4 = sns.heatmap(invertebrate_df, cbar=False, ax=ax4)
ax4.title.set_text('invertebrate_df')
g5 = sns.heatmap(virus_df, cbar=False, ax=ax5)
ax5.title.set_text('virus')

fig.tight_layout()
```
